# LOAD DATASET - multiple variables

```python
df = pd.read_csv("./mlr09.csv")
df.head()
```

```python
y = df["average_points_scored"].values
```

```python
df.iloc[:,:-1].head()
```

```python
X = df.iloc[:,:-1].values
```

```python
X[:5]
```

# RESCALED

```python
mu_X = np.mean(X, axis=0)
std_X = np.std(X, axis=0)

rescaled_X = (X - mu_X) / std_X
```

```python
rescaled_X[:5]
```

# VALIDATION

```python
lr.fit(rescaled_X, y)
```

```python
lr.coef
```

```python
lr.intercept
```
