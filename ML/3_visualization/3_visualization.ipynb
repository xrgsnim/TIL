### **Matplotlib**
- pyplot이라는 객체를 사용하여 데이터를 표시
- pyplot : 그래프들을 채워넣을 그림판 같은 느낌
- pyplot 객체에 그래프들을 쌓은 다음 flush


```python
import matplotlib.pyplot as plt
X = range(100)
Y = range(100)
plt.plot(X,Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]
```

```python
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)

plt.show()
```

**[ Figure & Axes ]**
- Matplotlib는 Figure안에 Axes로 구성
- Figure 위에 여러 개의 Axes를 생성(subplot 계속 그리는 느낌)

```python
fig = plt.figure()
# 사이즈 설정
fig.set_size_inches(8,4)
# figure 위에 subplot 넣음
ax_1 = fig.add_subplot(1,2,1) # 1번째 줄에 2개중 1번쨰
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1,Y_1)
plt.show()
```

**[ subplots ]**
- subplot의 순서를 grid로 작성

```python
# subplot의 구조 이해
fig = plt.figure()
fig.set_size_inches(5,5) # 사이즈 설정
# plt.style.use("ggplot") # 스타일 설정

ax = []
colors = ["b", "g", "r", "c", 'm', "y", "k"]
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i)) # 두 개의 plot 생성
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1,len(colors))]

    ax[i-1].plot(X_1, Y_1, c=c)
```

**[ set color ]**
- color 속성을 사용
- float : 흑백, rgb color, predefined color 사용

```python
plt.plot(X_1, Y_1, color="#CC0066") # 색깔 rgb로 맘껏 조정가능
plt.plot(X_2, Y_2, c="c")

plt.show()
```

**[ line style ]**
- https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html에서 어떤 스타일들이 있는지 참고

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.show()
```

**[ set title ]**
- pypllot에 .title함수 사용, figure의 subplot별 입력 가능
- latex타입의 표현도 가능(수식 표현 가능)

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.title("Two lines")
plt.show()
```

**[ set legned ]**
- legend 함수로 범례를 표시함, loc 위치 등 속성 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")
# fancybox : 범례 박스가 각지게 보이는
# loc : 범례 박스 위치
```

**[ grid ]**
- 격자 표시 가능
- plt.style.use("ggplot")과 같이 스타일을 조정해서 격자 표현도 가능

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")

# ls : 격자를 그리는 line style
# c : 투명도
# xlim : x축 최소,최대
# ylim : y축 최소,최대
plt.grid(True, lw=0.4, ls="--", c=".9")
plt.xlim(-10, 20)
plt.ylim(-10, 20)

plt.show()
```

**[ savefig ]**
- **.show()를 하는 순간 flush 되기 때문에 저장해도 보이지 않는다**
- 따라서 .show()하기 전에 저장 먼저 해줘야한다

```python
plt.savefig("test.png", c="a")
```

## scatter
- 산점도, .scatter()함수 사용,
- marker="~~" : scatter의 모양지정
- s : 데이터의 크기를 지정, 데이터의 크기 서로 비교가능

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
# c : color 
plt.scatter(data_1[:, 0], data_1[:, 1], c="b", marker="4")
plt.scatter(data_2[:, 0], data_2[:, 1], c="r", marker="o")

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
```

```python
# alpha : 투명도
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

### bar chart
- .bar()함수 사용 

```python
data = np.array([[5., 25., 50., 20. ],
        [4., 23., 51., 17. ],
        [6., 22., 52., 19. ],])

X = np.arange(4)
# plt.bar(X가 표현되는 x축상의 지점, y값)
plt.bar(X + 0.00, data[0], color="b", width=0.25)
plt.bar(X + 0.25, data[1], color="g", width=0.25)
plt.bar(X + 0.50, data[2], color="r", width=0.25)
plt.xticks(X+0.25, ("A", "B", "C", "D"))
plt.show()
```

```python
color_list = ["b", "g", "r"]
data_label = ["A", "B", "C"]
X = np.arange(data.shape[1])
```

```python
for i in range(3):
    plt.bar(
        X,
        data[i],
        bottom = np.sum(data[:i], axis=0),
        color = color_list[i],
        label = data_label[i]
    )
plt.legend()
plt.show()
```

```python
A = [5., 30., 45., 22.]
B = [5, 25, 50, 20]

X = range(4)
plt.bar(X, A, color="b")
# bottom : 다른 subplot의 시작지점
plt.bar(X, B, color="r", bottom=60)
plt.show()
```

```python
woman_pop = np.array([5, 30, 45, 22])
man_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

# barh : horizental bvar chart
plt.barh(X, woman_pop, color="r")
# '-' 붙여주면 반대로 출력됨
plt.barh(X, -man_pop, color="b")
plt.show()
```

### histogram

- bins="~~" : 몇개의 구간으로 나눌것인가

### boxplot
- Q1 : 0~25%
- Q2 : 25~50%
- Q3 : 50~75%
- Q4 : 75~100%
- IQR : Q1~Q3
- IQR내의 데이터를 주로 출현하는 데이터로 보고, 꺽쇠 밖의 값이 많을수록 outlier가 많다는 뜻
- 정규분포일수록 가운데 박스가 두툼해지고, 꺽쇠밖의 값이 많을수록 튀는 데이터가 많음

```python
data = np.random.randn(100, 5)
plt.boxplot(data)
plt.show()
```

## seaborn
- 통계적인 데이터를 visualization하기에 특화된 툴
- 기존 matplotlib에 기본 설정을 추가
- 복잡한 그래프를 간단하게 만들수 있음, matplotlib의 wrapper
- 간단한 코드 + 예쁜 결과물
- https://seaborn.pydata.org/tutorial.html에서 튜토리얼하면 좋음

### seaborn - basic plots
- matplotlib와 같은 기본적인 plot
- 손쉬운 설정으로 데이터 산출
- lneplot, scatterplot, countplot 등

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
# seaborn에서 기본제공하는 데이터
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
tips
```

```python
fmri
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="timepoint", y="signal", data=fmri)
# 연한부분 : x축에서 해당하는 데이터들의 분포,
# ex)timepoint가 5.0인 애들의 signal은 약 0.09에서 0.18까지 넓게 분포함
```

```python
sns.set_style("ticks")
sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
# 한 데이터셋 안에 numeric한 데이터와 categorical한 자료가 같이 있다면
# hue에 categorical한 칼럼을 넣어서 그 데이터에 따른 numeric자료들을 비교해볼 수 있다.
sns.lineplot(x = "timepoint", y = "signal", hue="event", data=fmri)
```

```python
# scatterplot : matplotlib의 scatter와 비슷
sns.scatterplot(x = "total_bill", y="tip", data=tips)
```

```python
# regplot : scatterplot + regression line : 추세선도 함께 표시
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
# scatterplot안에서 hue를 활용한 모습
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

#### countplot
- categorical한 데이터 이용할 때 주로 씀
- value_counts()와 같은 역할

```python
tips["smoker"].value_counts()
```

```python
sns.countplot(x = "smoker", data = tips)
```

```python
#bar chart보다 훨씬 간결한 코딩
sns.countplot(x = "smoker", hue="time", data = tips)
```

#### bar plot
- x축에는 category 데이터

```python
# 바의 끝부분 : 평균값
# 선의 길이 : 분포
sns.barplot(x="day", y="total_bill", data=tips)
```

#### distplot
- 분포를 나타낸다

```python
# kde : 분포의 곡선, 추정치로 출력
sns.displot(tips["total_bill"], bins=10, kde=True)
```

### seaborn - predefined plots
#### violinplot
- boxplot + 분포모양

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
```

#### swarmplot
-boxplot을 산점도 식으로 출력, 갯수로 그 경향성을 확인

```python
sns.swarmplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
```

```python
#catplot : 카테고리컬 플롯, 데이터를 분류함
# kind 파라미터를 정해서 다른 plot의 형태로 그릴수 있음
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

#### FacetGrid
- 행, 열 방향으로 서로 다른 조건을 적용하여 여러 개의 서브 플롯 제작
- 각 서브 플롯에 적용할 그래프 종류를 map() 메서드를 이용하여 그리드 객체에 전달

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill","tip")
```
